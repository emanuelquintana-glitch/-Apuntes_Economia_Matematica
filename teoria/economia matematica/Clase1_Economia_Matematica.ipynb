# La Naturaleza de la Economía Matemática

## ¿Qué es la Economía Matemática?

La economía matemática **no es una rama especializada** de la economía como las finanzas públicas o el comercio internacional, sino más bien **un método de análisis económico** donde el economista utiliza:

* **Símbolos matemáticos** para enunciar problemas
* **Teoremas matemáticos** para facilitar el razonamiento
* **Técnicas avanzadas** como álgebra matricial, cálculo, ecuaciones diferenciales

# La Naturaleza del Análisis Económico: Matemática vs. Literaria

El entendimiento de la Economía Matemática comienza por diferenciarla de la Economía "Literaria" (o no matemática). Esta distinción no radica en el tema de estudio, sino en el **instrumental lógico y el método de análisis** empleado.

---

## Diferencias Fundamentales en el Método de Análisis

| Aspecto Clave | Economía Matemática (Análisis Formal) | Economía Literaria (Análisis Intuitivo) |
| :--- | :--- | :--- |
| **Lenguaje y Comunicación** | Se basa en el uso de **símbolos, variables y ecuaciones** ($\frac{\partial U}{\partial x_i} = \lambda p_i$). La comunicación es universal y concisa. | Se fundamenta en el uso de **palabras, enunciados y descripciones narrativas**. La interpretación puede variar. |
| **Estructura Lógica** | Utiliza la **lógica deductiva formal** basada en **Teoremas, Axiomas y Proposiciones** (ej. $A \implies B$). La conclusión es inevitable si las premisas son verdaderas. | Emplea la **lógica literaria o verbal**, que a menudo se apoya en la **inducción**, la **analogía** y la **persuasión retórica**. |
| **Precisión y Rigor** | **Precisión es alta y cuantificable.** El razonamiento es explícito y susceptible de ser verificado paso a paso. Se buscan condiciones **necesarias y suficientes**. | **Precisión es variable.** El rigor depende de la habilidad del autor para articular las ideas. Se centra en la **viabilidad conceptual** y la plausibilidad. |
| **Suposiciones (Hipótesis)** | **Explícitas y formalizadas** en cada etapa del modelo (ej. convexidad de preferencias, rendimientos decrecientes). Esto permite el **Test de Sensibilidad**. | **Pueden ser implícitas** o diluidas en el texto. Esto dificulta la identificación precisa de los límites de validez del argumento. |
| **Naturaleza de la Conclusión** | **Conclusión cualitativa y cuantitativa** (ej. se puede resolver para un valor específico o demostrar la existencia de un equilibrio). | **Conclusión predominantemente cualitativa**, enfocada en la dirección de los efectos o la relación causal general. |

---

## Implicaciones Académicas

La **Economía Matemática** se distingue por su capacidad para:

1.  **Garantizar la Coherencia Interna:** Si un economista formula un modelo de forma matemática, el uso de teoremas garantiza que las conclusiones derivadas se desprendan lógicamente de las suposiciones iniciales.
2.  **Facilitar la Aplicación Empírica:** Al ser los modelos explícitos y cuantificados, la traducción a modelos econométricos para la contrastación estadística resulta directa.
3.  **Abordar la Complejidad:** Permite analizar sistemas con múltiples variables e interdependencias que serían intratables o propensos a errores en la lógica verbal.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# -------------------------------
# Funciones de utilidad Cobb-Douglas
# -------------------------------

def utilidad_general(bienes, alpha=0.5):
    """
    Función de utilidad Cobb-Douglas generalizada:
    U(x₁, x₂, ..., xₙ) = ∏ xᵢ^α
    """
    return np.prod(bienes**alpha)

def utilidad_2d(x, y, alpha=0.5):
    """
    Cobb-Douglas en 2 bienes: U(x, y) = x^α * y^α
    """
    return (x**alpha) * (y**alpha)

# -------------------------------
# Cálculos de ejemplo
# -------------------------------

bienes_2d = np.array([2, 3])
bienes_nd = np.array([2, 3, 4, 1.5, 2.5, 1.8])  # n = 6 bienes

util_2d = utilidad_2d(*bienes_2d)
util_nd = utilidad_general(bienes_nd)

print(f"Utilidad con 2 bienes: {util_2d:.2f}")
print(f"Utilidad con {len(bienes_nd)} bienes: {util_nd:.2f}")
print(f"La matemática permite generalizar de {len(bienes_2d)} a {len(bienes_nd)} bienes")

# -------------------------------
# Visualización
# -------------------------------

fig = plt.figure(figsize=(16, 5))
plt.suptitle("Generalización Matemática en Modelos de Utilidad", fontsize=15, y=1.05)

# ----- Gráfico 1: Contorno en 2D -----
ax1 = fig.add_subplot(131)
x = np.linspace(0.1, 5, 100)
y = np.linspace(0.1, 5, 100)
X, Y = np.meshgrid(x, y)
Z = utilidad_2d(X, Y)

contour = ax1.contourf(X, Y, Z, levels=30, cmap='plasma')
cbar = fig.colorbar(contour, ax=ax1)
cbar.set_label('Utilidad $U(x_1, x_2)$', rotation=270, labelpad=15)

ax1.set_xlabel('Bien $x_1$')
ax1.set_ylabel('Bien $x_2$')
ax1.set_title('Análisis Gráfico (2 Bienes)')
ax1.grid(alpha=0.2)

# ----- Gráfico 2: Superficie 3D -----
ax2 = fig.add_subplot(132, projection='3d')
Z = utilidad_2d(X, Y)

surface = ax2.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9, edgecolor='none')
ax2.set_xlabel('Bien $x_1$')
ax2.set_ylabel('Bien $x_2$')
ax2.set_zlabel('Utilidad $U$')
ax2.set_title('Superficie 3D: Visualización Extendida')

# ----- Gráfico 3: Comparativa dimensional -----
ax3 = fig.add_subplot(133)

n_bienes = np.arange(2, 11)
analisis_grafico = [1 if n <= 3 else 0 for n in n_bienes]
analisis_matematico = [1 for _ in n_bienes]

ax3.plot(n_bienes, analisis_grafico, 'r--o', label='Posible gráficamente')
ax3.plot(n_bienes, analisis_matematico, 'b-o', label='Posible con matemáticas')

ax3.axvline(3.5, color='gray', linestyle='--', linewidth=1)
ax3.text(3.6, 0.5, 'Límite gráfico (3D)', rotation=90, color='gray')

ax3.set_xticks(n_bienes)
ax3.set_yticks([0, 1])
ax3.set_yticklabels(['No', 'Sí'])

ax3.set_xlabel('Número de Bienes $n$')
ax3.set_ylabel('¿Análisis Posible?')
ax3.set_title('Comparativa: Matemáticas vs Visualización')
ax3.legend()
ax3.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Economía Matemática vs. Econometría: Diferencias Clave

La distinción entre Economía Matemática y Econometría reside fundamentalmente en su **objetivo primario** y su **método de validación**. Ambas son herramientas formales, pero abordan el análisis económico desde ángulos opuestos:

### 1. Economía Matemática (La Lógica Deductiva)

Se enfoca en la **formulación abstracta y el análisis teórico** de los modelos económicos.

| Aspecto | Foco |
| :--- | :--- |
| **Objetivo Primario** | Establecer la **coherencia interna** y la **validez lógica** de las teorías. |
| **Método Principal** | **Deducción Matemática**. Parte de un conjunto de supuestos ($A_1, A_2, \ldots$) para demostrar la existencia o propiedades de un equilibrio. |
| **Enfoque** | **Teórico y Analítico**. Se centra en las condiciones necesarias y suficientes para que ocurra un fenómeno. |
| **Producto Final** | Un **Teorema** o una **Proposición** formalmente demostrada. |

---

### 2. Econometría (La Evidencia Empírica)

Es el puente entre la teoría económica y la realidad empírica. Su propósito es **contrastar las teorías** y **cuantificar las relaciones** usando datos reales.

| Aspecto | Foco |
| :--- | :--- |
| **Objetivo Primario** | Proporcionar **validez empírica** a las teorías y estimar relaciones causales. |
| **Método Principal** | **Inducción Estadística**. Parte de la observación de datos muestrales para inferir y probar hipótesis sobre la población (ej., estimar el valor de un coeficiente $\beta$). |
| **Enfoque** | **Empírico y Cuantitativo**. Utiliza métodos estadísticos para manejar la **estocasticidad** (aleatoriedad) inherente a los datos económicos. |
| **Producto Final** | Una **Estimación** (un valor numérico) y sus **estadísticos de prueba** (valores $t$ o $p$). |

---

### Síntesis: Complementariedad

La **Economía Matemática** provee el **"esqueleto" lógico** de la teoría (el modelo $Y=f(X)$), mientras que la **Econometría** utiliza ese esqueleto para agregar la **"carne" de la realidad** (la estimación $\hat{Y} = \hat{\beta}X + \varepsilon$) y determinar si el modelo es plausible en la práctica.

```python
# Diagrama comparativo entre economía matemática y econometría
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Economía Matemática
aspectos_matematica = ['Teoría pura', 'Razonamiento deductivo', 'Símbolos matemáticos', 'Abstracción']
valores_matematica = [9, 8, 9, 8]

# Econometría
aspectos_econometria = ['Medición empírica', 'Prueba de hipótesis', 'Datos estadísticos', 'Validación teórica']
valores_econometria = [9, 9, 8, 8]

# Gráfico radial comparativo
angles = np.linspace(0, 2*np.pi, len(aspectos_matematica), endpoint=False).tolist()
angles += angles[:1]  # Cerrar el círculo
valores_matematica += valores_matematica[:1]
valores_econometria += valores_econometria[:1]

ax1 = plt.subplot(121, polar=True)
ax1.plot(angles, valores_matematica, 'o-', linewidth=2, label='Economía Matemática', color='blue')
ax1.fill(angles, valores_matematica, alpha=0.25, color='blue')
ax1.plot(angles, valores_econometria, 'o-', linewidth=2, label='Econometría', color='red')
ax1.fill(angles, valores_econometria, alpha=0.25, color='red')
ax1.set_xticks(angles[:-1])
ax1.set_xticklabels(aspectos_matematica)
ax1.set_ylim(0, 10)
ax1.set_title('Comparación: Enfoques y Características', size=14, pad=20)
ax1.legend(loc='upper right')

# Gráfico de complementariedad
ax2 = plt.subplot(122)
procesos = ['Formulación\nTeórica', 'Deducción\nMatemática', 'Prueba\nEmpírica', 'Validación\ny Aplicación']
etapas_matematica = [9, 9, 2, 3]
etapas_econometria = [3, 2, 9, 8]

x = np.arange(len(procesos))
ax2.plot(x, etapas_matematica, 'bo-', linewidth=3, markersize=8, label='Economía Matemática')
ax2.plot(x, etapas_econometria, 'ro-', linewidth=3, markersize=8, label='Econometría')
ax2.set_xlabel('Etapas del Análisis Económico')
ax2.set_ylabel('Nivel de Participación (1-10)')
ax2.set_title('Complementariedad de los Enfoques')
ax2.set_xticks(x)
ax2.set_xticklabels(procesos)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

# Aplicación Práctica: De la Teoría a la Implementación

## Fundamento teórico detrás del modelo

El modelo se basa en la **teoría de la elección del consumidor** en **microeconomía**, específicamente en la **maximización de la utilidad** sujeta a una **restricción presupuestaria**.

### El Problema de Maximización

El consumidor busca la combinación de bienes $\mathbf{x} = (x_1, x_2, \ldots, x_n)$ que le proporciona la mayor satisfacción (utilidad) dado su ingreso limitado y los precios de los bienes.

#### Función de Utilidad (Cobb-Douglas Logarítmica)

La función de utilidad utilizada es una forma logarítmica de la función **Cobb-Douglas**, que se caracteriza por ser **homotética** y **cóncava**, garantizando la existencia de una solución única y la **disminución de la utilidad marginal**.

$$U(\mathbf{x}) = \sum_{i=1}^n \alpha_i \ln(x_i)$$

Donde:
* $\mathbf{x}$: Vector de cantidades consumidas de los $n$ bienes.
* $x_i$: Cantidad consumida del bien $i$.
* $\alpha_i$: Peso o **preferencia** del consumidor por el bien $i$, con $\alpha_i > 0$.
* $\ln(\cdot)$: Logaritmo natural.

#### Restricción Presupuestaria

La suma del gasto en todos los bienes no puede exceder el ingreso total disponible, $I$.

$$\sum_{i=1}^n p_i x_i \leq I$$

Donde:
* $p_i$: Precio del bien $i$.
* $I$: Ingreso total (renta).
* La solución óptima siempre se encontrará en el límite, por lo que la restricción se trata como una igualdad.

---

### El Problema de Optimización (Formal)

El problema formal del consumidor es:

$$\max_{\mathbf{x}} \; U(\mathbf{x}) = \sum_{i=1}^n \alpha_i \ln(x_i)$$
$$\text{sujeto a } \sum_{i=1}^n p_i x_i \leq I$$

Este es un problema de optimización con restricción que se resuelve típicamente mediante el método de los **multiplicadores de Lagrange**.

### Solución Derivada (Demanda Marshalliana)

La solución a este problema (la **demanda Marshalliana** para cada bien $i$) revela que la **proporción del gasto** en el bien $i$ será igual al peso $\alpha_i$ relativo, asumiendo $\sum_{i=1}^n \alpha_i = 1$ para normalización (aunque la solución funciona para cualquier $\alpha_i > 0$).

$$\text{Gasto en bien } i: \quad p_i x_i^* = \frac{\alpha_i}{\sum_{j=1}^n \alpha_j} \cdot I$$

O la cantidad demandada óptima:

$$x_i^* = \frac{\alpha_i}{\sum_{j=1}^n \alpha_j} \cdot \frac{I}{p_i}$$

Esta formulación demuestra que la **elasticidad-ingreso** y la **elasticidad-precio cruzada** son unitarias para esta función de utilidad específica, lo que la hace un modelo simple y muy utilizado para ilustrar la teoría de la demanda.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# ------------------------------
# Función de utilidad Cobb-Douglas (logarítmica)
# ------------------------------
def utilidad_cobb_douglas(x, precios, ingreso, alphas):
    if np.any(x <= 0):
        return -np.inf  # Penaliza valores negativos o cero (log no definido)
    gasto_total = np.dot(precios, x)
    if gasto_total > ingreso:
        return -np.inf  # Penaliza si excede el ingreso
    return np.sum(alphas * np.log(x))

# ------------------------------
# Parámetros del consumidor
# ------------------------------
n = 4
precios = np.array([2.0, 3.0, 1.5, 4.0])
ingreso = 100
alphas = np.array([0.25, 0.35, 0.20, 0.20])

# Condición: suma de alphas = 1 para interpretar como participación marginal
assert np.isclose(np.sum(alphas), 1), "Las α deben sumar 1"

# Punto inicial: consumo igualitario
x0 = np.ones(n) * (ingreso / n / np.mean(precios))

# ------------------------------
# Optimización
# ------------------------------
def objetivo(x):
    return -utilidad_cobb_douglas(x, precios, ingreso, alphas)

# Restricción: gasto ≤ ingreso
restriccion = {
    'type': 'eq',
    'fun': lambda x: ingreso - np.dot(precios, x)
}

# Límites: no se puede consumir cantidad negativa
limites = [(1e-6, None) for _ in range(n)]

resultado = minimize(objetivo, x0, method='SLSQP', bounds=limites, constraints=[restriccion])

# ------------------------------
# Resultados
# ------------------------------
x_opt = resultado.x
gasto_opt = precios * x_opt
utilidad_max = -resultado.fun
porcentajes_gasto = 100 * gasto_opt / ingreso

# ------------------------------
# Visualización
# ------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
bienes = [f"$x_{i+1}$" for i in range(n)]

# Gráfico 1: Consumo óptimo
bars = ax1.bar(bienes, x_opt, color='cornflowerblue', alpha=0.8)
ax1.set_title("Asignación Óptima del Consumidor (Cobb-Douglas)", fontsize=13)
ax1.set_ylabel("Cantidad Óptima")
ax1.grid(axis='y', alpha=0.3)

for bar, valor in zip(bars, x_opt):
    ax1.text(bar.get_x() + bar.get_width()/2, valor + 0.2, f"{valor:.2f}", ha='center', fontweight='bold')

# Gráfico 2: Distribución del gasto óptimo
ax2.pie(porcentajes_gasto, labels=bienes, autopct='%1.1f%%', startangle=90,
        colors=plt.cm.Set3.colors[:n], wedgeprops={'edgecolor': 'white'})
ax2.set_title("Distribución del Gasto Total", fontsize=13)

plt.tight_layout()
plt.show()

# ------------------------------
# Impresión de resultados
# ------------------------------
print("Resultados del Modelo de Optimización del Consumidor:")
print("-" * 50)
print(f"Utilidad máxima alcanzada: {utilidad_max:.4f}")
print(f"Ingreso utilizado: {np.sum(gasto_opt):.2f} de {ingreso}")
print(f"Composición óptima (cantidades): {x_opt.round(2)}")
print(f"Gasto por bien: {gasto_opt.round(2)}")
```
