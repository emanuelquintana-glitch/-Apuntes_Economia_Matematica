```
# Ejemplo introductorio: Función de utilidad Cobb-Douglas
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros
alpha, beta = 0.6, 0.4
x = np.linspace(0.1, 10, 50)
y = np.linspace(0.1, 10, 50)
X, Y = np.meshgrid(x, y)

# Función de utilidad Cobb-Douglas
U = (X**alpha) * (Y**beta)

# Gráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, U, cmap='viridis', alpha=0.8)
ax.set_xlabel('Bien X')
ax.set_ylabel('Bien Y')
ax.set_zlabel('Utilidad U(x,y)')
ax.set_title('Función de Utilidad Cobb-Douglas\n$U(x,y) = x^{0.6} y^{0.4}$')
plt.tight_layout()
plt.show()
```
