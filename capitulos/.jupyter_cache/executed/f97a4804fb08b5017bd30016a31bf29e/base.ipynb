```python
# Ejemplo: Representacin de un modelo econmico simple
import matplotlib.pyplot as plt
import numpy as np


# Modelo de oferta y demanda
P = np.linspace(0, 20, 100)
Q_d = 100 - 3 * P  # Funcin de demanda
Q_s = 20 + 2 * P   # Funcin de oferta


plt.figure(figsize=(10, 6))
plt.plot(Q_d, P, 'r-', linewidth=2, label='Demanda: $Q_d = 100 - 3P$')
plt.plot(Q_s, P, 'b-', linewidth=2, label='Oferta: $Q_s = 20 + 2P$')


# Punto de equilibrio
P_eq = 16
Q_eq = 100 - 3 * P_eq
plt.plot([Q_eq], [P_eq], 'go', markersize=8, label=f'Equilibrio (Q={Q_eq}, P={P_eq})')


plt.xlabel('Cantidad (Q)')
plt.ylabel('Precio (P)')
plt.title('Modelo Bsico de Oferta y Demanda')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Ejemplo: Variables endgenas vs. exgenas en un modelo
import matplotlib.pyplot as plt


# Configuracin del grfico explicativo
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))


# Modelo con variable exgena
P_exogeno = 10  # Precio determinado externamente
Q_d = 100 - 3 * P_exogeno


ax1.bar(['Precio (P)', 'Cantidad (Q)'], [P_exogeno, Q_d], color=['red', 'blue'], alpha=0.7)
ax1.set_title('Variable Exgena\n(P determinada externamente)')
ax1.set_ylabel('Valor')
ax1.text(0, P_exogeno + 2, f'P = {P_exogeno}', ha='center', fontweight='bold')
ax1.text(1, Q_d + 2, f'Q = {Q_d}', ha='center', fontweight='bold')


# Modelo con equilibrio endgeno
# Resolviendo Q_d = Q_s para encontrar equilibrio
P_endo = 16  # Precio de equilibrio
Q_endo = 100 - 3 * P_endo


ax2.bar(['Precio (P)', 'Cantidad (Q)'], [P_endo, Q_endo], color=['green', 'purple'], alpha=0.7)
ax2.set_title('Variables Endgenas\n(Determinadas dentro del modelo)')
ax2.set_ylabel('Valor')
ax2.text(0, P_endo + 2, f'P = {P_endo}', ha='center', fontweight='bold')
ax2.text(1, Q_endo + 2, f'Q = {Q_endo}', ha='center', fontweight='bold')


plt.tight_layout()
plt.show()
```

```python
# Ejemplo: Relaciones vs. Funciones
import matplotlib.pyplot as plt
import numpy as np


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))


# Relacin (no funcin) - crculo
theta = np.linspace(0, 2*np.pi, 100)
x_rel = np.cos(theta)
y_rel = np.sin(theta)
ax1.plot(x_rel, y_rel, 'b-', linewidth=2)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Relacin (No Funcin)\n$x^2 + y^2 = 1$')
ax1.grid(True, alpha=0.3)
ax1.set_aspect('equal')


# Funcin - parbola
x_func = np.linspace(-2, 2, 100)
y_func = x_func**2
ax2.plot(x_func, y_func, 'r-', linewidth=2)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Funcin\n$y = x^2$')
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')


plt.tight_layout()
plt.show()
```

```python
# Ejemplo: Diferentes tipos de funciones econmicas
import matplotlib.pyplot as plt
import numpy as np


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))


x = np.linspace(0.1, 10, 100)


# Funcin constante (oferta monetaria fija)
ax1.axhline(y=100, color='red', linewidth=2)
ax1.set_title('Funcin Constante\n$M_s = 100$')
ax1.set_xlabel('Tiempo')
ax1.set_ylabel('Oferta Monetaria')
ax1.grid(True, alpha=0.3)


# Funcin lineal (funcin consumo)
C = 50 + 0.8 * x
ax2.plot(x, C, 'blue', linewidth=2)
ax2.set_title('Funcin Lineal\n$C = 50 + 0.8Y$')
ax2.set_xlabel('Ingreso (Y)')
ax2.set_ylabel('Consumo (C)')
ax2.grid(True, alpha=0.3)


# Funcin cuadrtica (funcin costo)
CT = 100 + 10*x + 0.5*x**2
ax3.plot(x, CT, 'green', linewidth=2)
ax3.set_title('Funcin Cuadrtica\n$CT = 100 + 10Q + 0.5Q^2$')
ax3.set_xlabel('Cantidad (Q)')
ax3.set_ylabel('Costo Total')
ax3.grid(True, alpha=0.3)


# Funcin racional (costo fijo promedio)
CFP = 100 / x
ax4.plot(x, CFP, 'purple', linewidth=2)
ax4.set_title('Funcin Racional\n$CFP = 100/Q$')
ax4.set_xlabel('Cantidad (Q)')
ax4.set_ylabel('Costo Fijo Promedio')
ax4.grid(True, alpha=0.3)


plt.tight_layout()
plt.show()
```
